```python
EGFR = pd.read_csv("../Limpieza_datos/Limpios/EGFR_limpio.csv")
```

```python
SRC = pd.read_csv("../Limpieza_datos/Limpios/SRC_limpio.csv")
```

```python
EGFR
```

```python
EGFR_col = EGFR[["Molecule ChEMBL ID", "Standard Value"]]
```

```python
SRC_col = SRC[["Molecule ChEMBL ID", "Standard Value"]]
```

```python
merge = pd.merge(EGFR_col, SRC_col, on="Molecule ChEMBL ID", suffixes=('_EGFR', '_SRC'))
```

```python
merge
```

```python
def inhibidor(IC50):
    if IC50 < 1000:
        return 1
    else:
        return 0
```

```python
merge["Actividad_EGFR"] = merge["Standard Value_EGFR"].apply(inhibidor)
merge["Actividad_SRC"] = merge["Standard Value_SRC"].apply(inhibidor)
```
