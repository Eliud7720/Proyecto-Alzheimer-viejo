```python
import pandas as pd
```

```python
inhibidores = pd.read_csv("Inhibidores_con_huella.csv")
```

```python
inhibidores.head()
```

```python
X = inhibidores.drop(columns=["Molecule ChEMBL ID", "Smiles", "ECFP"])
X
```

```python
import joblib
```

```python
modelo_cargado = joblib.load('RF_entrenado.joblib')
```

```python
y_predicha = modelo_cargado.predict(X)
```

```python
y_predicha
```

```python
df_predicho = pd.DataFrame(y_predicha, columns=["BBB"])
```

```python
df_predicho
```

```python
df_original = pd.read_csv("inhibidores.csv")
df_original
```

```python
df_concatenado = pd.concat([df_original, df_predicho], axis=1)
```

```python
df_concatenado
```

```python
probabilidades = modelo_cargado.predict_proba(X)
```

```python
probabilidades = probabilidades[:, 0]
```

```python
prob = pd.DataFrame(probabilidades, columns=["Probabilidad"])
```

```python
df_con = pd.concat([df_concatenado, prob], axis=1)
```
